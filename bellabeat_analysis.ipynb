#### Manipulate the data

The data has been manipulated in the following ways:

1. The datatype of the columns containing dates has been changed to datetime.
2. Some columns have been renamed for consistency.
3. New columns called `Day`have been added to various dataframes. These are weekdays extracted from the dates
4. Columns containing both date and time have been split into seperate date and time columns.

```python
# change the datatype of the column containing dates to datetime
df_daily_activity["ActivityDate"] = pd.to_datetime(df_daily_activity["ActivityDate"])
df_daily_activity.head()
```

```python
# rename the ActivityDay column to ActivityDate for consistency
df_daily_calories.rename(columns = {"ActivityDay": "ActivityDate"}, inplace = True)
df_daily_calories.head()
# change the datatype of the column containing dates to datetime
df_daily_calories["ActivityDate"] = pd.to_datetime(df_daily_calories["ActivityDate"])
# extract the day of the week from the date
df_daily_calories["Day"] = df_daily_calories["ActivityDate"].dt.day_name()
# view the first 5 entries in the dataframe
df_daily_calories.head()
```

```python
# rename the ActivityDay column to ActivityDate for consistency
df_daily_steps.rename(columns = {"ActivityDay": "ActivityDate"}, inplace = True)
# change the datatype of the column containing dates to datetime
df_daily_steps["ActivityDate"] = pd.to_datetime(df_daily_steps["ActivityDate"])
# extract the day of the week from the date
df_daily_steps["Day"] = df_daily_steps["ActivityDate"].dt.day_name()
df_daily_steps.head()
```

```python
# rename the SleepDay column to SleepDate for consistency
df_sleep_day.rename(columns = {"SleepDay": "SleepDate"}, inplace = True)
# change the datatype of the column containing date to datetime
df_sleep_day["SleepDate"] = pd.to_datetime(df_sleep_day["SleepDate"], format = "%m/%d/%Y %H:%M:%S %p")
# extract the day of the week from the date
df_sleep_day["Day"] = df_sleep_day["SleepDate"].dt.day_name()
df_sleep_day.head()
```

```python
# split date and time columns
new = df_hourly_calories["ActivityHour"].str.split(" ", n = 1, expand = True)
df_hourly_calories["time"] = new[1]
# change the datatype of the column containing time to datetime
df_hourly_calories["time"] = pd.to_datetime(df_hourly_calories["time"], format = "%I:%M:%S %p")
# df_hourly_calories["time"]
df_hourly_calories.head()
```

```python
# split date and time columns
new = df_hourly_steps["ActivityHour"].str.split(" ", n = 1, expand = True)
df_hourly_steps["time"] = new[1]
# change the datatype of the column containing time to datetime
df_hourly_steps["time"] = pd.to_datetime(df_hourly_calories["time"], format = "%I:%M:%S %p")
df_hourly_steps.head()
```

## Analyze

#### Summary statistics for daily activity

```python
# excluding the Id and ActivityDate column
df_dum = df_daily_activity.loc[:, ~df_daily_activity.columns.isin(["Id", "ActivityDate"])]
df_dum.describe()
```

Note that:
1. An average user logs 7637 steps per day.
2. An average user spends 990 minutes in a sedentary state per day.
3. An average user burns 2303 calories per day.

#### Average calories burnt by day of the week


```python
cats = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df_daily_calories_grouped =  df_daily_calories.groupby("Day")["Calories"].mean().reindex(cats)
df_daily_calories_grouped
```

#### Average steps by day of the week

```python
df_daily_steps_grouped =  df_daily_steps.groupby("Day")["StepTotal"].mean().reindex(cats)
df_daily_steps_grouped
```

#### Average total minutes asleep and total time in bed by day of the week

```python
df_sleep_day_grouped =  df_sleep_day.groupby("Day")[["TotalMinutesAsleep", "TotalTimeInBed"]].mean().reindex(cats)
df_sleep_day_grouped
```

#### Average calories burnt by time of the day

```python
df_hourly_calories_grouped =  df_hourly_calories.groupby([df_hourly_calories["time"].dt.hour])["Calories"].mean()
df_hourly_calories_grouped
```

#### Average steps by time of the day

```python
df_hourly_steps_grouped =  df_hourly_steps.groupby([df_hourly_steps["time"].dt.hour])["StepTotal"].mean()
df_hourly_steps_grouped
```

## Share

All data labels on the graphs have been formatted to 0 decimal places for better readibility.

```python
plt.figure(figsize=(10,6))
ax = df_daily_calories_grouped.plot.bar()
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.ylabel("Calories burnt")
plt.title("Average calories burnt by day of the week")
plt.show()
```

The most calories are burnt on Tuesday and the least calories are burnt on Thursday.

```python
plt.figure(figsize=(10,6))
ax = df_daily_steps_grouped.plot.bar()
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.ylabel("Steps taken")
plt.title("Average steps by day of the week")
plt.show()
```

The most steps are taken on Saturday and the least steps are taken on Sunday.

```python
df_sleep_day_grouped.head()
ax = df_sleep_day_grouped.plot.bar(stacked=True, figsize=(10,6))
for c in ax.containers:
    labels = [int(v.get_height()) if v.get_height() > 0 else '' for v in c]
    ax.bar_label(c, labels=labels, label_type='center')
plt.title("Time in bed and Time asleep by day of the week")
plt.ylabel("Time(minutes)")
plt.show()
```

The most difference between time in bed and time asleep is on Sunday and the least difference is Thursday.

```python
plt.figure(figsize=(10,6))
ax = df_hourly_calories_grouped.plot.bar()
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.xlabel("Time(hours")
plt.ylabel("Calories burnt")
plt.title("Average calories burnt by time of the day")
plt.show()
```

The most calories are burnt between 12:00-14:00 and 17:00-19:00. The least calories are burnt between 0:00-4:00.

```python
plt.figure(figsize=(10,6))
ax = df_hourly_steps_grouped.plot.bar()
for p in ax.patches:
    ax.annotate(str(int(p.get_height())), (p.get_x() * 1.005, p.get_height() * 1.005))
plt.xlabel("Time(hours)")
plt.ylabel("Steps taken")
plt.title("Average steps taken by time of the day")
plt.show()
```

The most steps are taken at 18:00 hours and the least steps are taken at 3:00 hours. Users are most active between 12:00-14:00 and 17:00-19:00. Users are least active between 0:00-5:00 hours.

```python
ax = df_daily_activity.plot.scatter(x  = "TotalSteps", y ="Calories", figsize=(10,6))
plt.xlabel("Total steps")
plt.ylabel("Calories burnt")
plt.title("Total steps vs. calories burnt")
plt.show()
```

As the total steps taken increase, the calories burnt increase as well. Majority of the users walk between 0-15000 steps and burn 0-3000 calories. Notice that there are calories burnt with 0 steps, this might be from other forms of exercise than walking.

```python
ax = df_sleep_day.plot.scatter(x  = "TotalTimeInBed", y ="TotalMinutesAsleep", figsize=(10,6))
plt.xlabel("Time in bed (minutes)")
plt.ylabel("Total time asleep (minutes)")
plt.title("Time in bed vs. time asleep")
plt.show()
```

As expected, there is a strongly positive correlation between time in bed and total time asleep.
